Filling NaNs with 0 and changing the dataype of price column

```python
audible_data['price']=audible_data['price'].str.replace(',',"")
```

```python
audible_data['price']=audible_data['price'].str.split('.').str.get(0)
```

```python
audible_data['price']=audible_data['price'].fillna(0)
```

```python
audible_data['author'] = audible_data['author'].str.replace(r'(?<=[a-z])(?=[A-Z])', ' ', regex=True)
audible_data['narrator'] = audible_data['narrator'].str.replace(r'(?<=[a-z])(?=[A-Z])', ' ', regex=True)
```

```python
audible_data['author'] = audible_data['author'].str.replace(r'(?<=\.)(?=[A-Z])', ' ', regex=True)
audible_data['narrator'] = audible_data['narrator'].str.replace(r'(?<=\.)(?=[A-Z])', ' ', regex=True)
```

```python
audible_data['stars']=audible_data['stars'].fillna(0)
```

```python
audible_data['stars']=audible_data['stars'].str.replace(',','')
```

```python
audible_data['stars']=audible_data['stars'].astype('int32')
```

```python
hours = audible_data['time'].str.extract(r'(\d+) hr', expand=False).fillna(0).astype(int)
minutes = audible_data['time'].str.extract(r'(\d+) min', expand=False).fillna(0).astype(int)
audible_data['total_mintues']=(hours*60)+minutes
```
