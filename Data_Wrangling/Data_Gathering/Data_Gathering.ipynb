# Working with CSV files

# 1. Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req=requests.get(url=url,headers=headers)
data=StringIO(req.text)
pd.read_csv(data)
```

# 9. Converters

```python
def rename(name):
    if name=='Royal Challengers Bangalore':
         return 'RCB'
    else:
         return name
    
rename('Royal Challengers Bangalore')
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

# Exporting Files

# 1. To_Csv

```python
import pandas as pd
df = pd.read_csv('deliveries.csv')
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index().to_csv('batsman_runs.csv',index=False)
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

# 2. To_Excel

```python
!pip install openpyxl
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df
```

```python
temp_df2 = df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    temp_df.to_excel(writer,sheet_name='Sheet_Name_1')
    temp_df2.to_excel(writer,sheet_name='Sheet_Name_2')
```

# To HTML

```python
df=pd.read_csv('deliveries.csv')
df.query('batsman_runs==6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count').to_html('sixesheatmap.html')
```

# To JSON

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack().to_json('ipl.json')
```
